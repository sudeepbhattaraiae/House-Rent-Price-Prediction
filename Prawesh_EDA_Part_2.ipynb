#### Steps for Data Cleaning

* Pick those columns that are relevant for the analysis and drop those that are not.
* Check for any null values
* Fill or drop the records with null values
* Find the percentage of the null records
* Filter the records according to the occurence

```python
#Getting the percentage of the null records
null_values=pd.DataFrame(house.isnull().sum(),columns=['null_sum'])
null_values=null_values[null_values.null_sum>0]
null_values['percentage']=(null_values.null_sum/len(house ))*100
null_values=null_values.sort_values(by='percentage',ascending=False)
null_values
```

```python
# Cleaning the region variable based on a forward slash ("/")
def region_forward_slash(x):
    if isinstance(x, str):
        tokens = x.split('/')
        return tokens[0].strip()
    else:
        return x  # Return the value as is if it's not a string (NaN)

house['region'] = house['region'].apply(region_forward_slash)
```

```python
def strip_last_two_uppercase_chars(string):
    if string is not None and len(string) >= 2 and string[-2:].isupper():
        return string[:-2].strip()
    elif string is not None:
        return string.strip()  # Adjusted to strip trailing whitespaces
    else:
        return string
```

#### Features engineering

```python
#fill missing values in 'cats_allowed' and 'dogs_allowed' columns with 0
house['cats_allowed'].fillna(0, inplace=True)
house['dogs_allowed'].fillna(0, inplace=True)

#convert 'cats_allowed' and 'dogs_allowed' columns to integers
house['cats_allowed'] = house['cats_allowed'].astype(int)
house['dogs_allowed'] = house['dogs_allowed'].astype(int)

#create a new column 'pets_allowed' in the pandas DataFrame 'house' that takes the value 1 if both 'cats_allowed' and 'dogs_allowed' are 1.
house['pets_allowed'] = (house['cats_allowed'] & house['dogs_allowed'])

# Drop the 'cats_allowed' and 'dogs_allowed' columns from the DataFrame
house.drop(['cats_allowed', 'dogs_allowed'], axis='columns', inplace=True)
```

#### Removing Outliers

```python
#checking how many values present
house.baths.value_counts()
```

```python
#drop rows with non-finite values in the 'baths' column
house = house.dropna(subset=['baths'])

#convert the values in the 'baths' column from float to integer data type
house["baths"] = house["baths"].astype(int)
```

```python
#checking how many values present
house.beds.value_counts()
```

```python
house
```

```python
outlier1 = ((house["beds"]>4) | (house["baths"]>4))
print("There is {} outlier".format(house[outlier1]["beds"].count()))
```

```python
house = house[~outlier1]
```

```python
house.shape
```

```python
outlier2 = ((house["sqfeet"]<120) | (house["sqfeet"]>5000) | (house["price"]<100) | (house["price"]>10000))
print("There is {} outlier".format(house[outlier2]["pets_allowed"].count()))
```

```python
house = house[~outlier2]
```

### DATA VISUALIZATION

```python
# Importing matplotlib for plotting
import matplotlib.pyplot as plt

# Importing seaborn for color palettes and styles
import seaborn as sns
```

```python
# Generate a color palette based on the number of unique values in 'type' column
colors = sns.color_palette('pastel', len(house['type'].unique()))

# Plot pie chart
house['type'].value_counts().plot.pie(explode=[0.05, 0.05,0.05,0.05, 0.05],autopct='%1.1f%%', colors=colors, shadow=True, figsize=(10, 8))

# Add legend with unique types
plt.legend(house['type'].unique(), loc="best")

# Add title
plt.title('Exploring the Diversity of House Types', fontsize=18)

# Show plot
plt.show()
```

The pie chart illustrates the diversity of house types based on the distribution of listings. Apartment properties dominate the listings,
followed by house, townhouses, condo, and duplex at least. This visualization provides a comprehensive overview of the types of properties
available in the housing market, facilitating better understanding and decision-making for buyers.

```python
import warnings
warnings.filterwarnings("ignore")

# Set a light colorful palette
colors = sns.color_palette('pastel')

# Create the bar plot using seaborn
sns.barplot(x=house["type"], y=house["price"], palette=colors)

# Set figure size
fig = plt.gcf()
fig.set_size_inches(15, 10)

# Add horizontal grid lines
plt.grid(axis='y', linestyle='--', alpha=0.2)

# Add title and labels
plt.title('Exploring Price Distribution Across Different House Types',fontsize=18)
plt.xlabel('House Type',fontsize=14)
plt.ylabel('Price in USD',fontsize=14)

# Add legend with unique types
plt.legend(house['type'].unique(), loc="upper right")

# Show the plot
plt.show()
```

The visualization depicts the distribution of house prices across different types, showcasing variations in pricing trends where the condos
and houses are pricey whereas apartments are cheaper comparitively.The bar plot offers a clear comparison of price ranges for various house
types.This summary aids in understanding the relationship between house types and their respective price distributions, facilitating informed decisions in the real estate market.

```python
# Create a figure and axis object with a specified size
fig, ax = plt.subplots(figsize=(16, 16))

# Group the data by square footage and calculate the mean price, then apply a rolling mean with window size 25
grouped = house.groupby("sqfeet")["price"].mean()
rollingGrouped = grouped.rolling(25).mean()

# Extract square footage and price data from the rolling mean
sqfeetData = rollingGrouped.index.values
priceData = rollingGrouped.values

# Set the title and labels for the plot
ax.set_title('Price per Square Footage', fontsize=18)
ax.set_xlabel("Square Footage", fontsize=14)
ax.set_ylabel("Price", fontsize=14)

# Plot the data as a line graph
ax.plot(sqfeetData, priceData)

ax.grid(axis='y', linestyle='--', alpha=0.1)

# Show the plot
plt.show()
```

The line graph illustrates the average price per square footage. As square footage increases, the price generally tends to rise, indicating a
positive correlation between these two variables.

```python
# Histograms and Boxplots for numerical features
numerical_features = ['price', 'sqfeet', 'beds', 'baths']
plt.figure(figsize=(12, 8))

# Plotting histograms for numerical features
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(house[feature], kde=True)
    plt.title(f'{feature.capitalize()} Distribution')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))

# Plotting boxplots for numerical features
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(house[feature])
    plt.title(f'{feature.capitalize()} Boxplot')

plt.tight_layout()
plt.show()
```

```python
# Countplots for categorical variables
categorical_features = ['type', 'laundry_options', 'parking_options']
plt.figure(figsize=(15, 5))

# Plotting countplots for categorical features
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(1, 3, i)
    sns.countplot(house[feature])
    plt.title(f'{feature.capitalize()} Countplot')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Scatter plots for numerical variables
plt.figure(figsize=(18, 5))

# Scatter plot of Price vs Sqfeet
plt.subplot(1, 3, 1)
sns.scatterplot(x='sqfeet', y='price', data=house)
plt.title('Price vs Sqfeet')

# Scatter plot of Price vs Beds
plt.subplot(1, 3, 2)
sns.scatterplot(x='beds', y='price', data=house)
plt.title('Price vs Beds')

# Scatter plot of Price vs Baths
plt.subplot(1, 3, 3)
sns.scatterplot(x='baths', y='price', data=house)
plt.title('Price vs Baths')

plt.tight_layout()
plt.show()
```

```python
# Heatmap for correlation between numerical features
plt.figure(figsize=(10, 8))

# Generating heatmap to visualize correlation between numerical features for the 'house' dataframe
sns.heatmap(house[numerical_features].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
